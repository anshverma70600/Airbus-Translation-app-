```python
from tkinter import *
from tkinter import ttk
from googletrans import Translator, LANGUAGES

def change(text="type",src="English",dest="Hindi"): 
    txt1 = text
    src1 = src
    dest1 = dest
    trans = Translator()
    trans1 = trans.translate(text,src=src1,dest=dest1)
    return trans1.text
    
def data():
    s = combo_src.get()
    d = combo_des.get()
    msg = s_txt.get(1.0,END)
    textget = change(text=msg,src=s,dest=d)
    d_txt.delete(1.0,END) 
    d_txt.insert(END,textget)





root = Tk()
root.title("Airbus Translator")
root.geometry("500x650")

lab_txt = Label(root,text="Airbus Translator",font=("Time New Roman",30,"bold"))
lab_txt.place(x=100,y=30,height=50,width=300)

t = Label(root,text="Enter text",font=("Time New Roman",20))
t.place(x=200,y=110,height=30,width=100)

frame = Frame(root).pack(side=BOTTOM)
s_txt = Text(frame,font=("Time New Roman",20),wrap=WORD)
s_txt.place(x=10,y=140,height=150,width=480)

list_text = list(LANGUAGES.values())

combo_src = ttk.Combobox(frame,value = list_text)
combo_src.place(x=10,y=310,height=40,width=170)

t = Label(root,text=" TO ==>",font=("Time New Roman",12))
t.place(x=200,y=310,height=30,width=100)

combo_des = ttk.Combobox(frame,value = list_text)
combo_des.place(x=320,y=310,height=40,width=170)

b_change = Button(frame,text="Proceed",relief=RAISED,command=data)
b_change.place(x=100,y=370,height=40,width=300)


t = Label(root,text="Translation",font=("Time New Roman",20))
t.place(x=200,y=420,height=30,width=100)

d_txt = Text(frame,font=("Time New Roman",20),wrap=WORD)
d_txt.place(x=10,y=450,height=150,width=480)






root.mainloop()
```
